```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
car = ['Car_Name','Year','Selling_Price','Present_Price','Driven_kms','Fuel_Type','Selling_type','Transmission','Owner']
for var in car:
    label_encoder = LabelEncoder()
    df[var+'_label_encoded'] = label_encoder.fit_transform(df[var])
df.drop(car,axis=1,inplace = True)
print("The encoded data\n:")
df
```

```python
df.corr(numeric_only = True)
```

```python
plot = plt.figure(figsize =(20,20))
plot = sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('CORELATION MATRIX')
plt.show()
```
