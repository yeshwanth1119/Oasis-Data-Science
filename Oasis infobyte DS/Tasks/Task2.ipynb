```python
ax=sns.countplot(x='Region', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

```python
df.columns = df.columns.str.replace(' ', '')
df.columns
```

```python
group = df.groupby(['Region'])['EstimatedUnemploymentRate(%)'].max()
group
```

```python
from itertools import groupby
group = df.groupby(["Region"])["EstimatedUnemploymentRate(%)"].aggregate("max").reset_index()
plt.figure(figsize=(12,8))
sns.pointplot(groupby['Region'].values,group['EstimatedUnemploymentRate(%)'].values,alpha=0.8)
plt.ylabel('Max rate', fontsize=12)
plt.xlabel('States', fontsize=12)
plt.title("Highest Rates of Unemployment by States", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```
